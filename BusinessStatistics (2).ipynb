## Measure of dispersion ##

Range, Variance, Standard deviation(Population)

```python
import numpy as np
```

```python
data = [1,2,3,4,4,5,6,7,7]
print("Population Range:", max(data) - min(data))     #Range
print("Population Mean:",np.var(data))                #Mean
print("Population Variance",np.std(data))             #Variance(ddof=0 by default > divide by n)
print("Population Std Dev", np.std(data))             # std dev (ddof=0 by default)
```

Range, Variance, Standard deviation(Sample)

```python
data = [1,2,3,4,4,5,6,7,7]
print("Sample Range:", max(data) - min(data))    #same as population
print("Sample Mean:",np.mean(data))              #Same mean
print("Sample Variance",np.var(data, ddof=1))    #Variance (unbiased > divide by n-1)
print("Sample Std Dev", np.std(data, ddof=1))    #std dev (unbiased)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis, mode
```

```python
# Generate synthetic dataset with some skewness and kurtosis
np.random.seed(42)
data = np.concatenate([
    np.random.normal(50,10,200),   # Normal distribution
    np.random.exponential(20,100)  # Right - skewed distribution
])

data = np.round(data, 2)  # Round to 2 decimal places
```

```python
# Create Dataframe
df = pd.DataFrame(data, columns=['Values'])
```

```python
# Print basic dataset info
print("Dataset Overview:")
print(f"Number of data points: {len(df)}")
print(f"First 5 values:\n{df.head()}\n")
```

```python
# Calculate basic statistics
mean_value = np.mean(data)
median_value = np.median(data)
mode_result = mode(data, keepdims=True)
mode_value = mode_result.mode[0]
data_range = np.ptp(data) # Peak-to-peak (range)
variance = np.var(data, ddof=1) # Sample variance
std_dev = np.std(data, ddof=1) # Sample Standard deviation
skewness = skew(data)
kurt = kurtosis(data)
```

```python
# Display calculated statistics
print("Descriptive Statistics:")
print(f"Mean: {mean_value:.2f}")
print(f"Median: {median_value:.2f}")
print(f"Mode: {mode_value:.2f}")
print(f"Range: {data_range:.2f}")
print(f"Variance: {variance:.2f}")
print(f"Standard Deviation: {std_dev:.2f}")
print(f"Skewness: {skewness:.2f}(Positive skew indicates longer right tail)")
print(f"Kurtosis: {kurt:.2f} (Positive value indicates sharper peak than normal distribution)\n")
```

## 1. What is Central Limit Theorem and give an example? ##

In probability theory, the central limit theorem (CLT) states that, under appropriate conditions, the distribution of a normalized version of the sample mean converges to a standard normal distribution. This holds even if the original variables themselves are not normally distributed.

Examplw: if you want to know the average weight of all apples in a huge orchard, you could take dozens of samples, weigh the apples in each sample, and calculate the average weight for each sample.

```python
import random
import statistics as stats
import matplotlib.pyplot as plt
```

```python
population = [random.expovariate(1) for _ in range(10000)]
```

```python
sample_size = 30
num_samples = 1000
```

```python
sample_means = []
for _ in range(num_samples):
    sample = random.sample(population, sample_size)
    sample_means.append(stats.mean(sample))
```

```python
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.hist(population, bins=50, color='purple', edgecolor = "Black", density=True)
plt.title('Original Distribution (Not Normal)')
plt.xlabel('Values')
plt.ylabel('Density')
```

```python
plt.subplot(1,2,2)
plt.hist(sample_means, bins=30, color="skyblue", edgecolor="black", density=True)
plt.title("sampling Distribution of Mean (CLT)")
plt.xlabel("Sample Means")
plt.ylabel("Density")
```

```python
plt.suptitle("Difference: Without CLT vs with CLT", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
print("Population Mean:", round(stats.mean(population), 3))
print("Population std deviation:", round(stats.pstdev(population), 3))
print("Mean of sample menas (CLT):", round (stats.mean(sample_means), 3))
print("std dev of sample means (CLT):", round(stats.pstdev(sample_means), 3))
```

##T-Test ##




```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t,ttest_1samp
```

**One-Sample t-Test Example (IQ & Medication)**

**Let's understand the t-test through a specific problem involving a one-sample t-test. In the population, the average IQ is 100. A team of researchers wants to test a new medication to determine if it has a positive, negative, or no effect on intelligence. Initially, the average IQ is 100. The researchers administer the medication to a sample of 30 participants, who have a mean IQ of 140 with a standard deviation of 20. The question is: did the medication affect intelligence? We will use a confidence interval of 95%, which corresponds to an alpha value of 0.05. Our goal is to analyze the sample data to conclude whether intelligence has increased or not due to the medication.**



```python
mu =100
n = 30
x_bar = 140
s = 20
alpha = 0.05
```

```python
df = n - 1
```

```python
t_stat = (x_bar - mu) / (s / np.sqrt(n))
```

```python
t_crit = t.ppf(1 - alpha / 2, df)
```

```python
p_value = 2 * (1 - t.cdf(abs(t_stat), df))
```

```python
print("One-Sample t-Test (IQ Medication)")
print(f"t statistic = {t_stat:.2f}")
print(f"Degrees of freedom = {df}")
print(f"Critical t values = ±{t_crit:.3f}")
print(f"P-value = {p_value:.5f}")
```

```python
x = np.linspace(-15, 15, 1000)
y = t.pdf(x, df)
```

```python
plt.figure(figsize=(8,5))
plt.plot(x, y, 'b')
plt.fill_between(x, y, where=(x < -t_crit) | (x > t_crit), color='red', alpha=0.3, label='Rejection region')
plt.axvline(t_stat, color='green', linestyle='--', label=f't = {t_stat:.2f}')
plt.axvline(-t_crit, color='black', linestyle='--')
plt.axvline(t_crit, color='black', linestyle='--')
plt.title("One-Sample t-Test for IQ Medication Study")
plt.legend()
plt.show()
```

**Example: Two-tailed One-Sample t-Test**

**A company claims the average working hours of employees per week is 40 hours. A labor union suspects it is different.They collect a random sample of 25 employees, and the results show: Sample mean = 42 hours Sample standard deviation = 4 hours Significance level = 0.05 (95% CI).**

```python
df = n - 1
```

```python
t_stat = (x_bar - mu) / (s / np.sqrt(n))
```

```python
t_crit = t.ppf(1 - alpha / 2, df)
```

```python
p_value = 2 * (1 - t.cdf(abs(t_stat), df))
```

```python
print("Example 2: Two-tailed One-Sample t-Test (Working Hours)")
print(f"t statistic = {t_stat:.2f}")
print(f"Degrees of freedom = {df}")
print(f"Critical t values = ±{t_crit:.3f}")
print(f"P-value = {p_value:.5f}")
```

```python
x = np.linspace(-4, 4, 1000)
y = t.pdf(x, df)
```

```python
plt.figure(figsize=(8,5))
plt.plot(x, y, 'b')
plt.fill_between (x, y, where=(x < -t_crit) | (x > t_crit), color='red', alpha=0.3, label='Rejection region')
plt.axvline(t_stat, color='green', linestyle='--', label=f't = {t_stat:.2f}')
plt.axvline(-t_crit, color='black', linestyle='--')
plt.axvline(t_crit, color='black', linestyle='--')
plt.title("Two-Tailed t-Test (Working Hours Example)")
plt.legend()
plt.show()
```

##Z-Score ##

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

**Example 1: Two-tailed Z Test (Heights)¶**


**Consider the following problem: The average height of all residents in a city is 168 centimeters.The population standard deviation is 3.9 cm.A doctor believes the mean height is different.The doctor measures the height of 36 individuals and finds the average height to be 169.5 cm. We will perform hypothesis testing at a 95% confidence interval to determine if there is enough evidence to reject the null hypothesis**

```python
mu =168
sigma =3.9
n = 36
x_bar = 169.5
alpha = 0.05
```

```python
z_stat = (x_bar - mu)/ (sigma / np.sqrt(n))
```

```python
z_crit = norm.ppf(1 - alpha/2)
```

```python
p_value = 2 * (1 - norm.cdf(abs(z_stat)))
```

```python
print("Example 1 (Two-tailed Test)")
print(f"Z statistic = {z_stat:.2f}")
print(f"Critical Z values = ±{z_crit:.2f}")
print(f"P-value = {p_value:.5f}")
```

```python
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)
```

```python
plt.figure(figsize=(8,5))
plt.plot(x, y, 'b')
plt.fill_between(x, y, where=(x < -z_crit) | (x > z_crit), color='red', alpha=0.3, label='Rejection region')
plt.axvline(z_stat, color='green', linestyle='--', label=f'Z = {z_stat:.2f}')
plt.axvline(-z_crit, color='black', linestyle='--')
plt.axvline(z_crit, color='black', linestyle='--')
plt.title("Two-tailed Z Test for Heights")
plt.legend()
plt.show()
```

**Example 2: One-tailed Z Test (Bulbs)¶**

**Consider a factory manufacturer of bulbs with: Average warranty of 5 years Standard deviation of 0.50 years Sample size of 40 bulbs A worker believes the bulbs malfunction in less than 5 years. The sample average is 4.8 years. We will test this hypothesis at a 98% confidence interval.**

```python
mu = 5.0
sigma = 0.50
n = 40
x_bar = 4.8
alpha =0.02
```

```python
z_stat = (x_bar - mu) / (sigma / np.sqrt(n))
```

```python
z_crit = norm.ppf(alpha)
```

```python
p_value = norm.cdf(z_stat)
```

```python
print("\nExample 2 (One-tailed Test)")
print(f"Z statistic = {z_stat:.2f}")
print(f"Critical Z value = {z_crit:.2f}")
print(f"P-value = {p_value:.5f}")
```

```python
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)
```

```python
plt.figure(figsize=(8,5))
plt.plot(x, y, 'b')
plt.fill_between(x, y, where=(x < z_crit) , color='red', alpha=0.3, label='Rejection region')
plt.axvline(z_stat, color='green', linestyle='--', label=f'z = {z_stat:.2f}')
plt.axvline(z_crit, color='black', linestyle='--')
plt.title("One-Sample Z Test for Bulbs")
plt.legend()
plt.show()
```
